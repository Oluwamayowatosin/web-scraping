```python
import pandas as pd
import numpy as np
```

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
my_url="https://www.flipkart.com/mixerjuicergrinders/pr?sid=j9e%2Cm38%2C7ek&otracker=nmenu_sub_Appliances_0_Mixer%2FJuicer%2FGrinder&otracker=nmenu_sub_TVs%20%26%20Appliances_0_Juicer%2FMixer%2FGrinder"
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
```

```python
page_soup = soup(page_html, "html.parser")
page_soup
```

```python
containers = page_soup.findAll("div", { "id": "container"})
print(len(containers))
print(soup.prettify(containers[0]))
```

```python
container = containers[0]
contents = container.findAll("a", {"class": "_1fQZEK"})
len(contents)
```

```python
prod_cont = contents[0]
```

```python
#let's scrape the product image and name
product_image = prod_cont.findAll("img", {"class": "_396cs4 _3exPp9"})
print(product_image[0]["src"])
print(product_image[0]["alt"])
```

```python
product_name = prod_cont.findAll("div", { "class": "_4rR01T"})
print(product_name[0].text)
```

```python
# Now let's have a look at the price
price = prod_cont.findAll("div", {"class": "_30jeq3 _1_WHN1"})
print(price[0].text)
```

```python
# Now let's have a look at the delivery
delivery = prod_cont.findAll("div", {"class": "_2Tpdn3"})
print(delivery[0].text)
```

```python
# Now let's have a look at what our CSV file has stored after the web scraping of Flipkart:
products_image = []
products_name = []
pricing = []
rating = []
review  = []
description = []
delivery  = []
```

```python
# Now let's have a look at what our CSV file has stored after the web scraping of Flipkart:

    
for cont in contents:
    # Now let's scrap the product image and name
    image = cont.findAll("img", {"class": "_396cs4 _3exPp9"})
    products_image.append(image[0]['src'])
    
    name = cont.findAll("div", {"class": "_4rR01T"})
    products_name.append(name[0].text)
    
    rate = cont.findAll("span", {"class": "_2_R_DZ"})
    if len(rate) ==0:
        rating.append(np.nan)
        review.append(np.nan) 
    else:
        split_val = rate[0].text.split()
        rating.append(split_val[0])
        review.append(split_val[3]) 
        
    describe = cont.findAll("ul", {"class": "_1xgFaf"})
    description.append(describe[0].text)
    
    price = cont.findAll("div", {"class": "_30jeq3 _1_WHN1"})
    pricing.append(price[0].text)
    
    deliv = cont.findAll("div", {"class": "_2Tpdn3"})
    delivery .append(deliv[0].text)
    
    
print(len(products_image))
print(len(products_name))
print(len(pricing))
print(len(review))
print(len(rating))
print(len(description))
print(len(delivery)) 
```
